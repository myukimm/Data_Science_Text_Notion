기계학습은 일반적으로 가르쳐주는 '교사'의 존재 여부에 따라 크게 지도 학습과 자율 학습으로 나누어진다. 그리고 보상과 처벌에 따른 학습을 하는 강화학습으로 분류할 수 있다.

**지도 학습** : 컴퓨터는 '교사'에 의해 주어진 예제와 정답(혹은 레이블)을 제공받는다. 지도학습의 목표는 입력을 출력에 매핑하는 일반적인 규칙을 학습하는 것이다. 예를 들어서 강아지와 고양이를 구분하는 문제라면 강아지와 고양이에 대한 이미지 데이터를 제공한 후에, 교사가 어떤 데이터가 강아지인지, 어떤 데이터가 고양이인지를 알려주는 것이다.

**자율 학습** : 외부에서 정답(레이블)이 주어지지 않고 학습 알고리즘이 스스로 입력에서 어떤 구조를 발견하는 학습이다. 흔히 비지도 학습이라고도 부르는 자율 학습을 사용하면 데이터에서 숨겨진 패턴을 발견할 수 있다. 

자율 학습의 대표적인 학습 방법은 클러스터링이다. 이 데이터의 경우 두 개의 큰 그룹으로 나누어 질 수 있는데, 이 규칙은 컴퓨터가 데이터를 보고 스스로 학습하는 것이다. 예를 들어서 구글 뉴스에서 비슷한 뉴스를 자동으로 그룹핑하는 것을 들 수 있다.

**강화학습** : 강화학습은 보상 및 처벌의 형태로 학습 데이터가 주어진다. 주로 차량 운전이나 상대방과의 경기 같은 동적인 환경에서 프로그램의 행동에 대한 피드백만 제공되는 경우이다. 예를 들어서 바둑에서 어떤 수를 두어서 승리하였다면 보상이 주어지는 방식이 될 수 있다.

### 지도학습

지도 학습은 주어진 입력-출력 쌍을 학습한 후에 새로운 입력값이 들어왔을 때, 합리적인 출력값을 예측하는 것이다. 즉 지도 학습은 입력(x)과 출력(y)이 주어질 때, 입력에서 출력으로의 매핑 함수 f( )를 학습하는 것이라 할 수 있다.

y = f(x)

지도 학습의 목표는 새로운 입력 데이터 (x)가 있을 때 해당 데이터에 대한 출력 변수값 y를 예측할 수 있도록 적절한 매핑 함수를 만드는 것이다.

간단한 예를 들어보자. (x, y) 형태의 입력 데이터로 점 (1,10), (2,20), (3,30), (4,40)들이 주어져 있다고 하자. 컴퓨터는 x 값에 대하여 y 값이 y = 10x의 방정식으로 표현될 수 있는 데이터라는 것을 아직 모르는 상태이다. 주어진 데이터 4개를 학습하여 학습이 끝난 후에 x = 5로 입력하면 컴퓨터가 y = 50이라는 답을 할 수 있도록 만들고 싶다. 이 경우 입력된 값을 바탕으로 컴퓨터가 스스로 이 입력을 설명할 수 있는 가장 좋은 함수를 찾는 것이 바로 지도학습이며, 이 문제는 지도학습 중에서도 회귀분석(regression)이라 할  수 있다.

#### 회귀문제

지도 학습 알고리즘에는 회귀 분석과 분류의 두가지 유형이 있다. 전자는 연속적인 값을 예측하고 후자는 객체를 어떤 카테고리 중의 하나로 예측한다. 예를 들어, 주택의 면적만을 고려하여 주택 가격을 예측하는 것은 회귀 분석이며, 주어진 이미지가 고양이인지 강아지인지를 판별하는 것은 분류이다.

회귀란 일반적으로 데이터들을 다차원 공간에 표시한 후에 이들 데이터들을 가장 잘 설명할 수 있는 직선이나 곡선을 찾는 문제라고 할 수 있다. 즉 y = f(x)에서 입력 x와 출력 y를 보면서 함수 f(x)를 예측하는 것을 회귀 기법이라고 한다. 예를 들어서 대상자들의 '키'를 x 좌표에 입력하고 이들의 '몸무게'를 y 좌표에 매핑시킨 후 이 상관관계를 가장 잘 설명하는 직선을 찾는 문제가 될 수 있을 것이다.

전통적인 선형회귀는 '기계학습'으로 생각하기에는 너무 단순하고 단순히 '통계'적 방법이라고 생각하는 사람들도 있다. 하지만 회귀 문제도 y = f(x)에서 입력 x에 대응되는 실수 y들이 주어지고 함수 f( )를 학습하는 것이므로 일종의 기계학습 문제로 생각할 수 있다.

데이터를 학습시킬 때, 데이터를 원형 그대로 사용하는 경우도 있지만 일반적으로는 데이터에서 어떤 특징을 추출하여 이것으로 학습시키고 테스트하게 된다. 그렇다면 특징은 무엇일까?

특징이란 관찰되는 현상에서 측정할 수 있는 개별적인 속성을 의미한다. y = f(x)에서는 x가 바로 특징이다. 기계학습에서 특징이라는 것은 학습의 결과를 결정하는 데에 영향을 미치는 입력 데이터라 할 수 있다.

##### 선형 회귀 분석

사이킷런(scikit-learn)을 사용하여 선형 회귀 분석에 대하여 간단히 살펴보자. 사이킷런은 파이썬 언어에서 기계학습을 수행하는 라이브러리이다. 이 라이브러리는 선형 회귀, k-nn 알고리즘, 서포트 벡터머신, 랜덤 포레스트, 그래디언트 부스팅, k-means등의 분류, 회귀, 클러스터링 알고리즘을 포함하고 있다.

선형 회귀는 임의의 변수 x와 이 변수에 따른 또 다른 변수 y와의 상관관계를 모델링하는 기법이다. 이 두 변수의 관계를 알아내거나 이를 이용하여 y가 없는 x값에 대해 y를 예측하는데 사용할 수 있다. 우선 가장 간단한 모델로 이차원 평면상에 있는 직선의 방정식을 생각해 보면, 이 직선의 방정식은 기본적으로 다음과 같다. 

y = ma + b

여기서 m은 직선의 기울기이고 입력 변수에 곱해지는 계수(coefficient)이다. b는 절편(intercept)이다. 기본적으로 선형 회귀 알고리즘은 데이터를 설명하는 가장 적절한 기울기와 절편값을 찾는다. x 변수는 데이터 특징이므로 변경할 수 없고 우리가 제어할 수 있는 값은 기울기(m)와 절편(b)이다. 기울기와 절편의 값에 따라 여러 개의 직선이 있을 수 있다. 기본적으로 선형 회귀 알고리즘은 데이터 요소에 여러 직선을 맞추어 본 후에 가장 적은 오류를 발생시키는 직선을 반환한다.

이 개념은 2개 이상의 변수가 있는 경우까지 확장될 수 있다. 이를 다중회귀분석이라고 한다. 예를 들어 주택의 면적, 침실 수, 해당 지역의 사람들의 평균 소득, 주택의 노후화 등을 기준으로 주택 가격을 예측해야 하는 시나리오를 생각해보자. 이 경우 종속 변수 p는 여러 독립 변수에 종속된다. p개의 변수가 포함된 회귀모델은 다음과 같이 나타낼 수 있다.

y(w,x) = w0 + w1x1 + ... wpxp

여기서 w와 x는 모두 벡터이고 w0을 제외한 w = (w1,w2,w3, ..., wp) 를 계수(coefficient), w0를 절편(intercept)이라고 한다. 이것은 사실 평면의 방정식이다. 2차원 공간에서 선형 회귀 모형은 직선이고 3차원에서는 평면이고, 3차원 이상에는 초평면이다. 

##### 선형회귀 사이킷런으로 구현

선형회귀를 위해 선형 모델 linear_model을 import 한 뒤에 LinearRegression() 함수를 통해 선형회귀 모델을 생성한다. 생성된 모델을 변수 regr로 받아오자.

```python
import numpy as np
from sklearn import linear_model

regr = linear_model.LinearRegression()
```

선형회귀를 위한 입력 데이터 X를 2차원 리스트에 넣도록 한다. 다음으로 y 변수를 초기화 하도록 하자. 그 후, regr.fit(X,y)와 같이 선형회귀 모델에 입력과 출력을 지정하면 된다.

```python
X = [[164], [179], [162], [170]] # 다중회귀에도 사용하도록 함
y = [53, 63, 55, 59]             # y = f(x)의 결과
regr.fit(X,y)
```

주의할 점은 **학습 데이터는 여러 변수에 대해서 다중회귀분석을 실시하기 위해서 반드시 2차원 배열**이어야 한다는 점이다. sklearn은 X의 각 항목을 스칼라 값이 아니고 다수의 종속 변수를 포함하는 벡터로 간주한다. 따라서 입력의 차원이 1차원인 경우에도 164가 아닌 [164]와 같은 배열 형태로 만들어야 한다.

선형 회귀 분석을 적용하려면 fit() 함수에 X와 y를 전달한다. sklearn은 보통 벡터값을 대문자로 표기하므로 X는 대문자를 사용하고 y는 스칼라 값이므로 소문자를 사용한다.

위의 값을 출력 함수로 나타내면 다음과 같다.

y = w0 + w1X1

여기서 계수는 (w1)이고 절편은 w0이다. 이것들이 이 직선의 방정식을 결정한다. 이 값을 확인하고 싶으면 선형회귀 모델이 가지고 있는 특성값 coef_와 intercept_를 확인하면 된다. 그리고 이 값들이 입력 X에 대해 y를 예측하는 데에 얼마나 적합한지는 score() 함수를 통해 확인할 수 있다. 아래와 같이 확인해 보자. 선형관계에 가까운 데이터를 사용했기 때문에 1에 가까운 0.9 정도의 값을 얻을 수 있다. 

```python
coef = regr.coef_             # 직선의 기울기
intercept = regr.intercept_   # 직선의 절편
score = regr.score(X,y)       # 학습된 직선이 데이터를 얼마나 잘 따르나

score
```

이제 이렇게 학습한 데이터를 바탕으로 새로운 입력이 주어지면 그에 맞는 결과를 예측할 수 있다. 이 예측은 우리가 학습을  통해 찾아 놓은 직선의 방정식을 통해 이루어질 것이다. 컴퓨터는 학습에 사용된 데이터를 보고 새롭게 들어오는 데이터도 이전 데이터와 같은 방식의 관계를 가질 것이라 예측하는 것이다. 값을 예측하는 함수는 predict()이다. 이 함수의 매개변수는 예측에 필요한 입력 데이터이고, 이 데이터는 선형회귀 학습을 할 때 사용한 입력 X와 같이 2차원 배열이다.

```python
input_data = [[180], [185]]
```

이것을 predict() 함수에 넘겨주면 예측결과를 얻을 수 있다.

```python
result = regr.predict(input_data)
print(result)
```

이제 얼마나 학습을 잘하고 있는 지 시각화를 해보자.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

regr = linear_model.LinearRegression()

X = [[164], [179], [162], [170]]  # 선형회귀의 입력은 2차원으로 만들어야 함
y = [53, 63, 55, 59]              # y = f(x)의 결과값

regr.fit(X, y)
# 학습 데이터와 y 값을 산포도로 그린다.
plt.scatter(X, y, color = 'black')

# 학습 데이터를 입력으로 하여 예측값을 계산한다.
y_pred = regr.predict(X)

# 학습 데이터와 예측값으로 선그래프로 그린다.
# 계산된 기울기와 y 절편을 가지는 직선이 그려진다.
plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()
```

이에 맞게 BMI 데이터만 가지고 오면 된다. 이는 세번째 열에 속하는 데 이것만 뽑아내면 1차원 데이터로 변형된다.

하지만, 선형 데이터는 입력값은 2차원 데이터가 되어야 한다. 

배열의 차원을 증가시키고 싶을 때는 넘파이의 newaxis 속성을 이용한다.

```python
X = diabetes.data[:, np.newaxis, 2]  # newaxis를 통하여 차원 1을 증가시킬 수 있다.
```

```python
regr.fit(X, diabetes.target)
print(regr.coef_, regr.intercept_)
```

그런데, 이렇게 학습을 마치고 얻는 계수와 절편을 이용하여 테스트를 해 볼 데이터가 존재하지 않는다. 그래서 원래의 데이터 중에서 일부는 학습에 사용하고, 남은 데이터를 가지고 학습결과가 예측을 잘 하는지 테스트해 보는 것이 좋을 것 같다.

학습을 위해서 전체 442개 중에서 80%만을 사용해 보자. 이 데이터로 학습하여 선형회귀 모델을 생성한 다음 나머지 20%를 테스트에 사용하려 한다.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2)
```

이제 x_train과 y_train과 같은 학습용 데이터를 이용하여 선형 회귀 모델로 학습을 수행한다. BMI 데이터만 사용하고 싶다면 앞 절에서 한 바와 같이 슬라이싱을 통해 데이터를 추출하고, 이것을 나누면 된다.

### KNN 분류

앞 절에서 다룬 데이터가 있을 때, 새로운 데이터가 입력된 경우를 생각해 보자. 새로운 데이터는 그림과 같이 녹색 별표로 표시되었다고 하자. 별표는 클래스 A와 클래스 B중 한 곳에 포함되어야 한다. 우리는 그 과정을 분류(classification)라고 부른다. 어떤 방법을 사용해야 할까?

가장 간단한 방법은 별표와 가장 가까운 이웃을 확인하는 것이다. 그림을 살펴보면 별표와 가장 가까운 이웃은 파란색 사각형이다. 따라서 별표는 파란색 사각형 클래스로 분류할 수 있을 것이다. 이와 같은 분류방법이 바로 k-nn 방법이다. 이 분류 방법은 가장 가까운 이웃에만 의존하기 때문에 최근접 이웃(Nearest Neighbor) 방법이라고 불린다.

2.jpg

하지만 가장 가까웃 이웃 하나만 보고 클래스를 판단하는 것은 문제가 있을 수 있다. 파란색 사각형이 가장 가깝지만 이보다 조금 떨어진 곳에 빨간색 원이 있다면 어떻게 될끼? 위의 그림에서 범위를 조금 확대하면 파란색 사각형보다 빨간색 원이 많다. 따라서 가장 가까운 것을 확인하는 것만으로는 충분하지 않다. 대신에 **우리는 가장 가까운 k개의 도형을 확인한다**. 그런 다음 그들 중 다수인 쪽으로 새로운 데이터를 속하게 한다. 

만약 k=3이라면 가장 가까운 3개의 도형을 취한다. 3개의 도형 중에서 2개는 빨간색 원(클래스 B)이고 1개는 파란색 사각형(클래스A)이다. 이 경우에 녹색 별은 빨간색 원의 그룹인 클래스 B에 추가되어야 한다. 그러나 만약 k=5라면 3개의 파란색 사각형과 2개의 빨간색 원이 속하게 된다. 따라서 새로운 도형은 파란색 사각형 클래스가 추가되어야 한다. 만약 k가 4라면 2개의 빨간색 동그라미가 2개 파란색 사각형보다 가까우니 전자를 택하는 게 합리적이다. **보통 k를 홀수로 취한다.** 

k-nn 방법은 특징 공간에 있는 모든 데이터에 대한 정보가 필요하다. 왜냐하면 가장 가까운 이웃을 찾기 위해 새로운 데이터에서 모든 기존 데이터까지의 거리를 확인해야 하기 때문이다. 따라서 데이터와 클래스가 많이 있다면, 많은 메모리 공간과 계산 시간이 필요하다는 단점이 있다. 반면에 알고리즘이 매우 단순해서 직관적이며, 사전 학습이나 특별한 준비 시간이 필요없다는 점이 장점이 될 것이다.

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
```

iris.data를 출력해 보면 4개의 열을 가지는 150개의 데이터를 볼 수 있는데, 이 배열의 각각의 열은 4개의 측정값(꽃받침 길이와 너비 및 꽃잎 길이와 너비)을 나타낸다. 향후 기계학습의 입력이 될 iris.data의 shape을 다음과 같이 다시 한 번 확인해 보자.

```python
iris.data.shape
```

iris 레이블은 0, 1, 2라는 이름으로 저장되어 있다. 이 레이블에서 0은 setosa, 1은 versicolor, 2는 virginica 종을 의미한다.

샘플 데이터가 가지는 4개의 특징 이름을 출력해 보자.

```python
iris.feature_names
```

여기서는 iris.data가 입력이며 iris.target이 학습 레이블이 될 것이다. 그리고 150개의 데이터 중에서 setosa versicolor, virginica 종이 각각 50개 임을 확인해 볼 수 있다.

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2)
```

이제 다음과 같이 KNeighborClassifier를 사용하여 학습과 테스트를 진행해 보자. 그리고 sklearn의 metrics를 사용하여 accuracy_score() 함수를 불러올 것이며 이 함수를 이용해서 k-nn의 예측치 y_pred와 y_test와의 차이를 확인해 볼 것이다.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

num_neigh = 1
knn = KNeighborsClassifier(n_neighbors = num_neigh)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
scores = metrics.accuracy_score(y_test, y_pred)
print('n_neighbors가 {0:d}일 때 정확도 : {1:.3f}'.format(num_neigh, scores))
```

위의 코드는 사이킷런 라이브러리에서 'KNeighborsClassifier' 클래스를 가져와서 knn이라는 인스턴스를 먼저 생성한다. 이때 사용되는 매개변수 'n_neighbors'는 k-nn 알고즘의 k 값을 결정한다. 여기서 k의 최적 값을 선택하는 것은 매우 중요한데, 이 코드에서는 k = 6를 사용한다. 일반적으로 k의 값이 증가함에 따라 정확도가 증가지만 너무 커질 경우 다시 감소하는 경우가 있다.

### 과적합과 과소적합

기계학습에서 흔히 볼 수 있는 가능성은 과적합(overfitting)이다. 과적합이란 학습하는 데이터에서는 성능이 뒤어나지만 새로운 데이터에 대해 적용했을 때(이를 일반화라 한다). 성능이 잘 나오지 않는 모델을 만들어내는 일이다. 과적합이 나오는 이유는 데이터에 섞인 잡음까지 학습하기 때문이다. 이와 반대로 데이터에 대한 학습이 지나치게 대충 이루어져 학습 데이터와 다른 데이터 모두에 대해 예측을 제대로 하지 못하는 모델을 만드는 일은 과소적합(underfitting)이라고 한다.

아래의 그림을 살펴 보자. 첫번째 그림은 데이터의 분포와 이 분포를 설명하는 선이 실제 데이터와 너무 큰 차이를 보이고 있다. 이러한 모델의 경우 새로운 데이터를 잘 예측하지 못하므로 과소적합되었다고 할 수 있다. 반면 가운데 그림은 데이터의 분포와 이를 설명하는 선이 비교적 잘 일반화 되어 좋은 적합을 보인다고 할 수 있다. 가장 오른쪽의 그림은 데이터의 분포와 이를 설명하는 선이 지나치게 맞추어져 있어서 새로운 데이터에 대한 예측 성능이 오히려 떨어질 수 있다. 이러한 적합이 과적합이다.
